```python
import toyplot
import pandas
```

## Load dataset and print the first 5 rows

```python
data = pandas.read_csv(toyplot.data.cars.path)
data.head()
```

## The X and Y axes will be for MPG and Horsepower respectively.  The Origin and Cylinders will be represented using other marker features.

### Obtain the unique cases in Origin and Cylinders

```python
cylinder_options = sorted(pandas.unique(data['Cylinders']).astype(int))
origin_options = sorted(pandas.unique(data['Origin']).astype(int))

print(cylinder_options)
print(origin_options)
```

### For both Origin and Cylinders generate arrays of marker shapes and marker colors respectively that correspond to each sample in the dataset. Then check to confirm there is a shape and color for each row in the dataset.

```python
colors = []
shapes = []

for c in data['Cylinders']:
    if c == 3:
        colors.append("black")
    elif c == 4:
        colors.append("red")
    elif c == 5:
        colors.append("blue")
    elif c == 6:
        colors.append("purple")
    elif c == 8: 
        colors.append("green")
        
for o in data['Origin']:
    if o == 1:
        shapes.append("o")
    elif o == 2:
        shapes.append("s")
    elif o ==3:
        shapes.append("d")
        
print(data.shape)
print(len(colors))
print(len(shapes))
```

### Display MPG vs Horsepower with shapes and colors representing the Origin and Cylinders

```python
canvas = toyplot.Canvas(1000,1000)
axes = canvas.cartesian(label = "MPG vs. Horsepower vs. Origin vs. Cylinders", xlabel="MPG", ylabel="Horsepower")
mark = axes.scatterplot(data['MPG'], data['Horsepower'], color=colors, marker=shapes, size = 10)
```
