# Data Visualization in Python: Introduction to Toyplot

## Toyplot Background

### What is ***toyplot***?

***Toyplot*** is a data visualization library in python.  It is user-friendly and intuitive, as well as a good place to start with data visualization in python compared to the more complex *matplotlib* library.  With ***toyplot*** you can:

- Develop custom interactive, animated plots that embrace the unique capabilities of electronic publishing and support reproducibility
- Create the best possible data graphics "out-of-the-box", maximizing data ink and minimizing chart junk
- Provide and clean, minimalist interface that scientists and engineers will love

## General Conventions

### Canvas

In toyplot, a *canvas* is simply the space that your visualization will be created on.  This can be as small or large as you desire.  For the most part your visualizations will be scaled to the size of the *canvas* that you define.

### Axes

*Axes* refer to the "coordinate system(s)" that will be added to the canvas for your data visualization(s).

### Mark

*Marks* refer to the actual visualization(s) that is generated for or added to your coordinate system(s).

```python
import toyplot

# Define canvas with width and height in inches.  Can also use centimeters (cm) or pixels (no units just integers)

canvas = toyplot.Canvas("10in", "5in")

# Define Cartesian coordinate axes

axes = canvas.cartesian()

# Add mark

mark = axes.scatterplot([10, 1],[5, 2], size=10)
```
