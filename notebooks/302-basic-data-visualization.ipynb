# Data Visualization in Python: Basic Data Visualization

## Histograms and Bar Charts

### Histograms

*Histograms* refer to graphical representations that display data by way of bars to show the frequency of numerical data.

### Bar Charts

*Bar charts* refer to pictorial representations that use bars to compare different categories of data.

From a programming perspective, there isn't much of difference between visualizing a histogram and bar chart.  Because of this both are implemented the same way using toyplot, which is using the **bars** mark.

```python
import numpy
import toyplot
```

```python
data = numpy.random.randint(0, 10, (10,10))
hist = numpy.histogram(data)
print()
print("The raw data is:")
print()
print(data)
print()
print("The histogram for the data is:")
canvas = toyplot.Canvas(500,500)
axes = canvas.cartesian()
mark = axes.bars(hist)
print()
print("The (stacked) bar chart of the data is:")
print()
canvas = toyplot.Canvas(500,500)
axes = canvas.cartesian()
mark = axes.bars(data)
```

## Line Charts

*Line charts* are charts that display information as a series of data points connected by a straight line.  They are generated in toyplot using the **plot** mark.

```python
x = numpy.arange(0, 10)
y = x**2
canvas = toyplot.Canvas(1000, 500)
axes = canvas.cartesian()
mark = axes.plot(x,y)
```

## Scatter Plots

*Scatter plots* are mathematical diagrams the using Cartesian coordinates to display values for two variables for a set of data.  Though scatter plots typically only show data for 2 variables, there are ways to increase the number of variables displayed by adding an additional axis, using colors, marker size, marker symbol, etc.  They are generated in toyplot using the **scatterplot** mark.

```python
data = numpy.random.randint(0, 10, (10,2))

canvas = toyplot.Canvas(1000, 500)
axes = canvas.cartesian()
mark = axes.scatterplot(data[:,0], data[:,1], size=10)
```
